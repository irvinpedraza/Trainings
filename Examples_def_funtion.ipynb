#funcion 3, correccion del interes compuesto tomando en cuenta la inflacion

interes = [12,11.5,11.2,19.7,11.9]
inflacion = [5,4.2,4.1,4,4.6]

def ICadj(m,i,f):
  registro = [m]
  for j in range(len(interes)):
    M = registro[j]*(1+((i[j]-f[j])/100))
    registro.append(M)
  return registro


```python
ICadj(25000,interes,inflacion)
```

```python
#Funtion 4, agregar contactos
contactos = []

def agregar_contacto():
  nombre = input("Ingrese el nombre del contacto: ")
  apellido = input("Ingrese el Apellido del contacto: ")
  edad = input("Ingrese la edad del contacto: ")
  telefono = input("Ingrese el telefono del contacto: ")

  #crear diccionario
  nuevo_contacto = {
      'nombre': nombre,
      'apellido': apellido,
      'edad': edad,
      'telefono': telefono
  }

  contactos.append(nuevo_contacto)
  print("Contacto agregado")
```

```python
agregar_contacto()
```

```python
print(contactos)
```
